```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Grepsr/Dataset - Junior Data QA Analyst.csv', encoding='utf-8')
```

```python
grouped = df.groupby('post_code')
```

```python
with open('/content/drive/MyDrive/Grepsr/Task 1/manifest.txt', 'w', encoding='utf-8') as f:
    for post_code, group in grouped:
        unique_dobs = group['dob'].unique()
        total_unique_dob = len(unique_dobs)
        total_records = len(group)
        f.write(f"post_code: {post_code}\n")
        f.write(f"dob: {', '.join(map(str, sorted(unique_dobs)))}\n")
        f.write(f"total_unique_dob: {total_unique_dob}\n")
        f.write(f"total_records: {total_records}\n\n")
```
